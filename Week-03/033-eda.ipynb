```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('nutrition-food.csv', encoding='UTF-16')
```

```python
df
```

```python
df.columns
```

```python
plt.figure(figsize=(10,5))
plt.plot(df[' Calories'], df[' Protein (g)'], marker='o', color='pink', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.title('Calories vs Protein Line Plot', fontsize=20, color='green')
```

```python
random_arr = np.random.choice([-1, 0, 1], df.shape[0])
random_arr = np.random.choice([-1, 0, 1], df.shape[0], p=[0.1, 0.3, 0.6])
random_arr
```

```python
df['cat'] = random_arr
df.columns
```

```python
df[' Fiber (g)'] = random_arr
df.columns
```

```python
plt.figure(figsize=(10,5))
plt.bar(x=df[' Fiber (g)'].unique(), height=df[' Fiber (g)'].value_counts().to_list(), color=['#0f5a3e', '#f8b195', '#f67280'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of Categories', fontsize=20, color='green')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(df[' Calories'], df[' Fat (g)'], marker='o', color='red', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel(' Calories')
plt.ylabel(' Fat (g)')
plt.title(' Fat (g) vs Calories', fontsize=20, color='green')

plt.subplot(1,2,2)
plt.plot(df[' Carb. (g)'], df[' Protein (g)'], marker='o', color='pink', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel(' Carb. (g)')
plt.ylabel(' Protein (g)')
plt.title(' Carb. (g) vs  Protein (g) Plot', fontsize=20, color='green')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df[' Carb. (g)'], df[' Protein (g)'], marker='o', color='Silver', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.fill_between(df[' Carb. (g)'], df[' Protein (g)'], color='#8cabc2')
plt.xlabel(' Carb. (g)')
plt.ylabel(' Protein (g)')
plt.title(' Carb. (g) vs  Protein (g) Line Plot', fontsize=20, color='#b7cede')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[[' Calories', ' Fat (g)']], palette=['#767bb2', '#dde8ee'])
plt.xlabel('Categories')
plt.ylabel(' Fat (g)')
plt.title('Calories vs  Fat (g)', fontsize=20, color='#8cabc2')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=' Calories', data=df.reset_index(), palette=['#767bb2', '#8cabc2'])
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('calories', fontsize=20, color='#b7cede')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(df[[' Calories', ' Fat (g)']], palette=['#c5a1a1', '#d6c875'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Calories vs  Fat (g)', fontsize=20, color='Red')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = ['#c5a1a1' if val > 0 else '#d4b288' for val in df[' Calories']] 
ax.scatter(df[' Calories'], df[' Fat (g)'], df[' Carb. (g)'], color=colors)
ax.set_xlabel(' Calories')
ax.set_ylabel(' Fat (g)')
ax.set_zlabel(' Carb. (g)')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import pdist
```

```python
df_plot = df.drop('Unnamed: 0', axis=1)
df_plot
```

```python
plt.figure(figsize=(10,5))
link = linkage(df_plot, 'ward')
dendrogram(link, color_threshold=5, truncate_mode='lastp', p=10)
plt.xticks(rotation=90)
plt.title('Dendrogram Plot', fontsize=20, color='red')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.clustermap(df_plot)
plt.show()
```

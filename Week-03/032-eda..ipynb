```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df= pd.read_csv('nutrition-drinks.csv')
df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Calories'], df['Fat (g)'], marker='o', color='pink', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Calories')
plt.ylabel('Fat')
plt.title('Calories vs Fat Line Plot', fontsize=20, color='green')
```

```python
random_arr = np.random.choice([-1, 0, 1], df.shape[0])
random_arr = np.random.choice([-1, 0, 1], df.shape[0], p=[0.1, 0.3, 0.6])
random_arr
```

```python
df['cat'] = random_arr
df.columns
```

```python
df['Fiber (g)'] = random_arr
df.columns
```

```python
plt.figure(figsize=(10,5))
plt.bar(x=df['Fiber (g)'].unique(), height=df['Fiber (g)'].value_counts().to_list(), color=['#0f5a3e', '#f8b195', '#f67280'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of Categories', fontsize=20, color='green')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(df['Carb. (g)'], df['Calories'], marker='o', color='red', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Total Carbohydrates')
plt.ylabel('Calories')
plt.title('Carbohydrates vs Calories', fontsize=20, color='green')

plt.subplot(1,2,2)
plt.plot(df['Protein'], df['Sodium'], marker='o', color='pink', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Protein')
plt.ylabel('Sodium')
plt.title('Protein vs Sodium Plot', fontsize=20, color='green')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Protein'], df['Sodium'], marker='o', color='Silver', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.fill_between(df['Protein'], df['Sodium'], color='#8cabc2')
plt.xlabel('Protein')
plt.ylabel('Sodium')
plt.title('Protein vs Sodium Line Plot', fontsize=20, color='#b7cede')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Calories', data=df.reset_index(), palette=['#767bb2', '#8cabc2'])
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('calories', fontsize=20, color='#b7cede')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = ['#c5a1a1' if val > 0 else '#d4b288' for val in df['Calories']] 
ax.scatter(df['Protein'], df['Sodium'], df['Calories'], color=colors)
ax.set_xlabel('Protein')
ax.set_ylabel('Sodium')
ax.set_zlabel('Calories')
plt.show()
```

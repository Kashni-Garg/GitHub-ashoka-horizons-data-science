```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df= pd.read_csv('all-menu.csv')
df
```

```python
df.dtypes
```

```python
arr=['Vitamin A (% DV)', 'Calcium (% DV)', 'Iron (% DV)']

for col in df.columns:
    if col in arr:
        df[col]=df[col].str.replace('%','')
        df[col].astype(float)
```

```python
arr=['Vitamin A (% DV)', 'Calcium (% DV)', 'Iron (% DV)']

for col in df.columns:
    if col in arr:
        df[col].head(5)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Calories'], df['Caffeine (mg)'], marker='o', color='pink', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Calories')
plt.ylabel('Caffeine (mg)')
plt.title('Calories vs Caffeine (mg) Line Plot', fontsize=20, color='green')
```

```python
random_arr = np.random.choice([-1, 0, 1], df.shape[0])
random_arr = np.random.choice([-1, 0, 1], df.shape[0], p=[0.1, 0.3, 0.6])
random_arr
```

```python
df['cat'] = random_arr
df.columns
```

```python
df['Vitamin A (% DV)'] = random_arr
df.columns
```

```python
df['Vitamin A (% DV)']
```

```python
plt.figure(figsize=(10,5))
plt.bar(x=df['Vitamin A (% DV)'].unique(), height=df['Vitamin A (% DV)'].value_counts().to_list(), color=['#0f5a3e', '#f8b195', '#f67280'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of Categories', fontsize=20, color='green')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(df[' Total Carbohydrates (g) '], df['Calories'], marker='o', color='red', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Total Carbohydrates (g)')
plt.ylabel('Calories')
plt.title('Carbohydrates vs Calories', fontsize=20, color='green')

plt.subplot(1,2,2)
plt.plot(df['Calories'], df['Caffeine (mg)'], marker='o', color='pink', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.xlabel('Calories')
plt.ylabel('Caffeine (mg)')
plt.title('Calories vs Caffeine Plot', fontsize=20, color='green')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Calories'], df['Caffeine (mg)'], marker='o', color='Silver', markersize='10.0', markerfacecolor='blue', linestyle=':')
plt.fill_between(df['Calories'], df['Caffeine (mg)'], color='#8cabc2')
plt.xlabel('Calories')
plt.ylabel('Caffeine')
plt.title('Calories vs Caffeine Line Plot', fontsize=20, color='#b7cede')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Calories', ' Total Carbohydrates (g) ']], palette=['#767bb2', '#dde8ee'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Calories vs Carbohydrates', fontsize=20, color='#8cabc2')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Calories', data=df.reset_index(), palette=['#767bb2', '#8cabc2'])
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('calories', fontsize=20, color='#b7cede')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(df[['Calories', ' Total Carbohydrates (g) ']], palette=['#c5a1a1', '#d6c875'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Calories vs Carbohydrates', fontsize=20, color='Red')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = ['#c5a1a1' if val > 0 else '#d4b288' for val in df['Calories']] 
ax.scatter(df[' Protein (g) '], df['Trans Fat (g) '], df['Saturated Fat (g)'], color=colors)
ax.set_xlabel(' Protein (g) ')
ax.set_ylabel('Trans Fat (g) ')
ax.set_zlabel('Saturated Fat (g)')
plt.show()
```

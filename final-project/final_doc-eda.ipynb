This project forcused on analysing the Insulin and Glucose levels of People suffering from Diabetes and people who are not from a set of provided data.

# Step by Step Explanation 
#   Importing libraries

- !pip install pandas
- !pip install numpy
- import pandas as pd
- import numpy as np

- - Firstly the the pandas and numpy libraries were installed and imported to perform funcions on the given .csv  file 

- df= pd.read_csv('diabetes.csv')

- - Variable 'df' is use to read the 'diabetes.csv' file 

# Basic functions are performed on the file 

- df.dtypes : finding out the data type of the variables stored in each column
- df.columns : finding out the name of different columns 
- df.head(5) : printing the first five rows of the file
- df.tail(5) : printing the last 5 rows of the file 
- df.describe() : summary of the data 

# Analysing the data 

- df.isnull().sum() : Sum of the null values present in each columns
- df.isna().sum() : Sum of n/a values present in each columns 

# Categorizing the data 

- with_diabetes=df[df['Outcome']==1] : as the variable name suggests, it stores the data of people with diabetes
- without_diabetes=df[df['Outcome']==0] : here the data of people not suffering from diabetes is stored

# Visualization of data 

- !pip install matplotlib
- !pip install seaborn 
- import matplotlib.pyplot as plt 
- import seaborn as sns

- - matplotlib and seaborn libraries are installed and imported for visualization of data. 

# Plotting the data 

- plt.figure(figsize=(10,5))
- sns.scatterplot(with_diabetes[['Glucose', 'Insulin']], palette=['#25573e', '#de7850'])
- plt.xlabel('Glucose and Insulin levels', color='#80cba4')
- plt.ylabel('Count', color='#80cba4')
- plt.title('People with Diabetes', fontsize=20, color='Red')

- - : a scatterplot of the Glucose and Insuline levels of people with diabetes 


- plt.figure(figsize=(10,5))
- sns.scatterplot(without_diabetes[['Glucose', 'Insulin']], palette=['#25573e', '#de7850'])
- plt.xlabel('Glucose and Insulin levels', color='#80cba4')
- plt.ylabel('Count', color='#80cba4')
- plt.title('People without Diabetes', fontsize=20, color='Red')

- - :a scatterplot of the glucose and insuline levels of people without diabetes 


# Insights 

- From the scatterplots we can conclude that the people suffering for diabeted have a higher level of Glucose and Insulin compared to the people with out it.
